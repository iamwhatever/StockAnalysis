```python
searchUrlLeading = "https://www.amazon.com/s?k=toddler+shoes+boys&page="

for i in range(1, 10):
    searchUrl = searchUrlLeading + str(i)
    print(searchUrl)
    
    getPageFromURL(searchUrl)
```

```python
import requests

def getPageFromURL(url): 
    response = requests.get(url)
    
    print(response.text)
```

```python
import requests

user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'
headers = {'User-Agent': user_agent}

# Download the page using requests
r = requests.get("https://www.amazon.com/s?k=toddler+shoes+boys&page=2", headers=headers)

print(r.text)
```
